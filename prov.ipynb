```python
grouped_df = {desk: list(zip(list(df[df['description']==desk]['block']), list(df[df['description']==desk]['quantity']))) for desk in df['description'].unique()}
grouped_df
```

```python
data_fr = []
for key in grouped_df.keys():
    data_fr.append([key, ''])
    for item in grouped_df[key]:
        data_fr += [item]
    data_fr.append(['', ''])
pd.DataFrame(data_fr)
```
